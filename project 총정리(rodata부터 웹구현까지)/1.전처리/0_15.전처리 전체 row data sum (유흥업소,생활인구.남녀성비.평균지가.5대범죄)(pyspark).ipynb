```python
crime.rename(columns={"지역코드":"지역구"},inplace=True)
crime = crime.groupby("지역구").mean()
```

```python
kara.rename(columns={"지역코드":"지역구"},inplace=True)
kara.rename(columns={"지번주소":"유흥업소개수"},inplace=True)
kara = kara.groupby("지역구").count()
kara

```

```python
den = den.groupby("지역코드").mean()
den.reset_index(inplace=True)
den.rename(columns={"지역코드":"지역구"},inplace=True)
den
```

```python
floating = floating.groupby("지역코드").mean()
floating.reset_index(inplace=True)
floating.rename(columns={"지역코드":"지역구"},inplace=True)
floating
```

```python
# single = single.groupby("지역코드").mean()
# single.reset_index(inplace=True)
# single.rename(columns={"지역코드":"지역구"},inplace=True)
single.rename(columns={"합계":"여성1인가구합계"},inplace=True)
```

```python
gender = gender.groupby("지역코드").mean()
gender.reset_index(inplace=True)
gender.rename(columns={"지역코드":"지역구"},inplace=True)
gender
```

```python
cctv.rename(columns={"개수":"cctv개수"},inplace=True)
```

```python
cctv = cctv.groupby("지역코드").mean()
cctv.reset_index(inplace=True)
cctv.rename(columns={"지역코드":"지역구"},inplace=True)
police = police.groupby("지역코드").mean()
police.reset_index(inplace=True)
police.rename(columns={"지역코드":"지역구","sum(경찰관수)":"경찰관수","sum(total)":"파출소개수"},inplace=True)
```

```python
price = price.groupby("지역코드").mean()
price.reset_index(inplace=True)
price.rename(columns={"지역코드":"지역구"},inplace=True)
```

```python
a = crime.merge(cctv,on="지역구")
b = a.merge(police,on="지역구")
c = b.merge(price,on="지역구")
d = c.merge(gender,on="지역구")
e = d.merge(single,on="지역구")
f = e.merge(floating,on="지역구")
g = f.merge(den,on="지역구")
final = g.merge(kara,on="지역구")
```

```python
del final["ID"]
del final["index_x"]
del final["index_y"]
```
