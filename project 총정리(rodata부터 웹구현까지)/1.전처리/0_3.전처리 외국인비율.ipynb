```python
a = df4['연도'] == 2017
b = df4['연도'] == 2018
c = df4['연도'] == 2019
```

```python
_2017 = df4[a]
_2017["외국인"] = _2017.등록_외국인.str.replace(',','').astype('int64')
_2018 = df4[b]
_2018["외국인"] = _2018.등록_외국인.str.replace(',','').astype('int64')
_2019 = df4[c]
_2019["외국인"] = _2019.등록_외국인.str.replace(',','').astype('int64')
# _2017["외국인밀도"] = _2017["등록_외국인"]/_2017["면적"]
```

```python
del _2017["등록_외국인"]
del _2018["등록_외국인"]
del _2019["등록_외국인"]
```

```python
_2017["외국인밀도"] = _2017["외국인"] / _2017["면적"]
_2018["외국인밀도"] = _2018["외국인"] / _2018["면적"]
_2019["외국인밀도"] = _2019["외국인"] / _2019["면적"]
```

```python
fin = pd.concat([_2017,_2018,_2019])
```

```python
sns.set(rc = {'figure.figsize':(20,20)})
sns.set(font='Malgun Gothic', 
        rc={"axes.unicode_minus":False},
        style='darkgrid')
sns.lmplot(x="5대범죄",y='외국인밀도',hue="연도",data=fin)
```
