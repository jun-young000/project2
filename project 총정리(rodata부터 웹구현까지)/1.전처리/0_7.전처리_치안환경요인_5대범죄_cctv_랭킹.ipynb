```python
# 데이터 불러오기 
cctv_seoul=pd.read_csv('../[0]data/서울시_CCTV.csv',index_col=0)
```

```python
cctv_seoul=cctv_seoul.reset_index()
```

```python
cctv_seoul=cctv_seoul.set_index(['년도','지역구'])
```

```python
cctv_seoul2017=cctv_seoul.loc[2017]
cctv_seoul2018=cctv_seoul.loc[2018]
cctv_seoul2019=cctv_seoul.loc[2019]
```

```python
cctv_seoul2017
cctv_seoul2018
cctv_seoul2019
```

```python
cctv_seoul1=pd.merge(cctv_seoul2017,cctv_seoul2018,on="지역구")
```

```python
cctv_seoul=pd.merge(cctv_seoul1,cctv_seoul2019,on="지역구")
```

```python
cctv_seoul.rename(columns = {'합계_x':'cctv합계(2017)','합계_y':'cctv합계(2018)','합계':'cctv합계(2019)'},inplace = True)
```

```python
cctv_seoul.info()
```

```python
cctv_seoul['cctv합계'] = (cctv_seoul['cctv합계(2017)']+cctv_seoul['cctv합계(2018)']+cctv_seoul['cctv합계(2019)']/3).round()
```

```python
cctv_seoul=cctv_seoul[['cctv합계']]
```

```python
crime_ranking=pd.read_csv('./5대범죄발생율합계.csv',index_col=0)
```

```python
cctv_seoul=cctv_seoul.reset_index()
```

```python
cctv_seoul['지역구'][(cctv_seoul['지역구'] == '중 구')] = '중구'
```

```python
# cctv_seoul=cctv_seoul.replace()
```

```python
cctv_seoul
```

```python
cctv_seoul=cctv_seoul.set_index(['지역구'])
```

```python
cctv_crime=pd.merge(crime_ranking,cctv_seoul,on="지역구")
```

```python
cctv_crime
```

```python
len(cctv_crime)
```

```python
cctv_crime.to_csv('cctv_범죄(준영오빠).csv')
```

```python
cctv_crime['5대범죄발생대비cctv개수'] = ((cctv_crime['5대범죄합계']/cctv_seoul['cctv합계'])*100)
```

```python
cctv_crime =cctv_crime.sort_values(by='5대범죄발생대비cctv개수',ascending=False)
```

```python
cctv_crime=cctv_crime.round()
```

```python
cctv_crime['순위']=cctv_crime['5대범죄발생대비cctv개수'].rank(ascending=False)
```

```python
cctv_crime=cctv_crime.astype(int)
```
