```python
CCTV = pd.read_csv('../[0]data/서울시_CCTV_년도별.csv',encoding='euc-kr',header=1,thousands=',')
CCTV
```

```python
CCTV.rename(columns={CCTV.columns[0]:"지역구"},inplace=True)
```

```python
CCTV=CCTV.drop(index=0)
```

```python
CCTV['2017']= CCTV['총계'] - CCTV['2021년'] - CCTV['2020년'] - CCTV['2019년'] - CCTV['2018년']
CCTV['2018']= CCTV['총계'] - CCTV['2021년'] - CCTV['2020년'] - CCTV['2019년'] 
CCTV['2019']= CCTV['총계'] - CCTV['2021년'] - CCTV['2020년'] 
```

```python
CCTV=CCTV[['지역구','2017','2018','2019']]
```

```python
CCTV
```

```python
CCTV_2017=CCTV[['지역구','2017']]
CCTV_2018=CCTV[['지역구','2018']]
CCTV_2019=CCTV[['지역구','2019']]
```

```python
CCTV_2017
CCTV_2018
CCTV_2019
```

```python
def func(x) :
    if x >=0  :
        return 2017
    else :
        return "none"
CCTV_2017["년도"] = CCTV_2017["2017"].apply(lambda x : func(x))
```

```python
CCTV_2017
```

```python
CCTV_2017.rename(columns = {'2017' : '합계'},inplace = True)
CCTV_2017
```

```python
def func(x) :
    if x >=0  :
        return 2018
    else :
        return "none"
CCTV_2018["년도"] = CCTV_2018["2018"].apply(lambda x : func(x))
```

```python
CCTV_2018
```

```python
CCTV_2018.rename(columns = {'2018' : '합계'},inplace = True)
CCTV_2018
```

```python
def func(x) :
    if x >=0  :
        return 2019
    else :
        return "none"
CCTV_2019["년도"] = CCTV_2019["2019"].apply(lambda x : func(x))
```

```python
CCTV_2019
```

```python
CCTV_2019.rename(columns = {'2019' : '합계'},inplace = True)
CCTV_2019
```

```python
cctv_total= pd.concat([CCTV_2017,CCTV_2018,CCTV_2019])
```

```python
cctv_total
```

```python
cctv_total.set_index(['년도','지역구'],inplace=True)
```

```python
cctv_total
```

```python
cctv_total.to_csv('../[0]data/서울시_CCTV.csv')
```
