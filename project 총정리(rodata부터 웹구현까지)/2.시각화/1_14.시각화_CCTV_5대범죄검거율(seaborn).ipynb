### 1.시각화 5대범죄 발생 검거 데이터 

```python
# 데이터 불러오기 

crime_rates=pd.read_csv('../[0]data/서울시_5대범죄_발생_검거.csv',index_col=0)
crime_rates.head(40)
crime_rates.tail(40)
```

```python
crime_rates['살인검거']=pd.to_numeric(crime_rates['살인검거'], errors='coerce')
```

```python
crime_rates.head(40)
crime_rates.tail(40)
```

```python
crime_rates=crime_rates.fillna(0)
```

```python
crime_rates.head(40)
crime_rates.tail(40)
```

```python
crime_rates=crime_rates.sort_index()
```

```python
crime_rates
```

```python
crime_rates.reset_index(drop=False,inplace=True)
```

```python
crime_rates
```

```python
crime_rates.set_index(['기간','지역구'],inplace=True)
```

```python
crime_rates
```

```python
crime_rates["강도검거율"] = crime_rates["살인검거"] / crime_rates["살인발생"] * 100
crime_rates["강간검거율"] = crime_rates["강도검거"] / crime_rates["강도발생"] * 100
crime_rates["살인검거율"] = crime_rates["강간강제추행검거"] / crime_rates["강간강제추행발생"] * 100
crime_rates["절도검거율"] = crime_rates["절도검거"] / crime_rates["절도발생"] * 100
crime_rates["폭력검거율"] = crime_rates["폭력검거"] / crime_rates["폭력발생"] * 100
```

```python
crime_rates
```

```python
crime_rates=crime_rates[["강도검거율","강간검거율","살인검거율","절도검거율","폭력검거율"]]
```

```python
crime_rates.head(40)
crime_rates.tail(40)
```

```python
# crime_rates.index
crime_2017=crime_rates.loc[2017]
crime_2018=crime_rates.loc[2018]
crime_2019=crime_rates.loc[2019]
```

```python
cctv_2017=pd.read_csv('../[0]data/서울시_cctv_2017.csv',index_col=0)
cctv_2018=pd.read_csv('../[0]data/서울시_cctv_2018.csv',index_col=0)
cctv_2019=pd.read_csv('../[0]data/서울시_cctv_2019.csv',index_col=0)
```

```python
cctv_2017
cctv_2018
cctv_2019
```

```python
cctv_2017=cctv_2017[['합계']]
cctv_2018=cctv_2018[['합계']]
cctv_2019=cctv_2019[['합계']]
```

```python
cctv_2017
cctv_2018
cctv_2019
```

```python
crime_cctv_2017 = pd.merge(crime_2017,cctv_2017,on='지역구')
crime_cctv_2018 = pd.merge(crime_2018,cctv_2018,on='지역구')
crime_cctv_2019 = pd.merge(crime_2019,cctv_2019,on='지역구')
```

```python
crime_cctv_2017
crime_cctv_2018
crime_cctv_2019
```

```python
crime_cctv_2017.to_csv('../[0]data/5대범죄검거율_2017.csv')
crime_cctv_2018.to_csv('../[0]data/5대범죄검거율_2018.csv')
crime_cctv_2019.to_csv('../[0]data/5대범죄검거율_2019.csv')
```

```python
crime_cctv_2017.rename(columns = {'합계':'CCTV(2017)'},inplace = True)
crime_cctv_2018.rename(columns = {'합계':'CCTV(2018)'},inplace = True)
crime_cctv_2019.rename(columns = {'합계':'CCTV(2019)'},inplace = True)
```

```python
crime_cctv_2017
crime_cctv_2018
crime_cctv_2019
```

```python
#CCTV 갯수와 5대범죄 검거율에 대한 상관 관계


sns.pairplot(crime_cctv_2019,y_vars=["CCTV(2019)"], x_vars=["강도검거율", "강간검거율", "살인검거율", "절도검거율", "폭력검거율"], 
             size=3,kind="reg").set(ylim=(0,5000))
plt.savefig('../[0]data/5대범죄검거율과 cctv_2017(시각화).png')

sns.pairplot(crime_cctv_2018, y_vars=["CCTV(2018)"], x_vars=["강도검거율", "강간검거율", "살인검거율", "절도검거율", "폭력검거율"], 
             size=3, kind="reg").set(ylim=(0,5000))
plt.savefig('../[0]data/5대범죄검거율과 cctv_2018(시각화).png')

sns.pairplot(crime_cctv_2017, y_vars=["CCTV(2017)"], x_vars=["강도검거율", "강간검거율", "살인검거율", "절도검거율", "폭력검거율"], 
             size=3, kind="reg").set(ylim=(0,5000))
plt.savefig('../[0]data/5대범죄검거율과 cctv_2019(시각화).png')
```
