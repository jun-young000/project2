```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

```python
rank = pd.read_csv("E:\data_science\DB\랭킹전처리\인구환경순위.csv")
```

```python
del rank["Unnamed: 0"]
```

```python
rank.sort_values("인구환경순위",ascending=False,inplace=True)
```

```python
rank_top = rank.head(3)
rank_bot = rank.tail(3)


```

```python
rank_sum = pd.concat([rank_top,rank_bot],axis=0)
```

```python
colors = ["#4E6BA6","#85BFF2"]
palette = sns.color_palette(colors)
sns.set(font='Malgun Gothic', 
        rc={"axes.unicode_minus":False},
        style='darkgrid')
sns.barplot(y="지역구",x="인구환경순위",data=rank_sum,palette = palette)
plt.savefig("pop_top_bot.png")
```
