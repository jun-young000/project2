```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

```python
living = pd.read_csv("주거.csv")
arrest = pd.read_csv("경찰관수 1인당 검거율_랭킹.csv")
```

```python
del arrest["년도"]
del arrest["경찰관수"]
del arrest["치안시설총수"]
del arrest["발생합계"]
del arrest["살인발생"]
del arrest["강도발생"]
del arrest["강간강제추행발생"]
del arrest["절도발생"]
del arrest["강간강제추행검거"]
del arrest["절도검거"]
del arrest["폭력검거"]
```

```python
del arrest["폭력발생"]
del arrest["검거합계"]
del arrest["강도검거"]
del arrest["5대범죄검거율"]
del arrest["강도검거율"]
del arrest["절도검거율"]
del arrest["폭력검거율"]
del arrest["살인검거율"]
del arrest["강간강제추행검거율"]
```

```python
del arrest["살인검거"]
del arrest["Unnamed: 0"]
```

```python
arrest["랭크"] = arrest["랭크"].astype(int)
```

```python
del living["Unnamed: 0"]
```

```python
living["범죄비율"] = living["5대범죄"]/living["주거비율"]
```

```python
living["순위"] = living["범죄비율"].rank().astype(int)
```

```python
del living["주거비율"]
del living["5대범죄"]
del living["범죄비율"]
```

```python
arrest.sort_values("랭크")
```

```python
total = arrest.merge(living,on="지역구")
```

```python
total["주거환경순위"] = total.랭크 + total.순위
```

```python
del total["랭크"]
del total["순위"]
```

```python
import pandas as pd
import numpy as np
import folium

import json
from folium.features import DivIcon
```

```python
geo_path = ('./seoul_municipalities_geo_simple.json')

get_str = json.load(open(geo_path, encoding='utf-8'))
```

```python
seoul_map = folium.Map(location = [37.5502, 126.982],zoom_start = 10.5)
```

```python
folium.Choropleth(
    geo_data = get_str,
    data = fin,
    columns = ['지역구', '최종순위'],
    fill_color='OrRd',
    key_on='properties.name',
    ).add_to(seoul_map)
```

```python
text = '중구'
circle_lat = 37.5650
circle_lon = 126.999


folium.map.Marker(
    [circle_lat, circle_lon],
    icon=DivIcon(
        icon_size=(10,6),
        icon_anchor=(0,0), 
        html='<div style="font-size: pt; font-weight:bold;">%s</div>' % text,
        )
    ).add_to(seoul_map)
```
